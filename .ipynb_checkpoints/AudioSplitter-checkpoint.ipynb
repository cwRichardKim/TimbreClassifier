```python
import numpy, scipy, matplotlib.pyplot as plt, pandas, librosa
from IPython.display import Audio
```

```python
x, fs = librosa.load('chopin.wav')
print x.shape
print fs

chopin_short, fs2 = librosa.load('Chopin short.mp3')
bohemian_short, fs3 = librosa.load('Bohemian Rhapsody 300ms.mp3')
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
chopin_stft = librosa.stft(chopin_short)
test = librosa.logamplitude(np.abs(chopin_stft)**2,ref_power=np.max)
librosa.display.specshow(test,y_axis='log', x_axis='time')
```

```python
bohemian_stft = librosa.stft(bohemian_short)
librosa.display.specshow(librosa.logamplitude(np.abs(bohemian_stft)**2,ref_power=np.max),y_axis='log', x_axis='time')
```
