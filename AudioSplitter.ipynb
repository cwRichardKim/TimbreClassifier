```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
chopin_stft = librosa.stft(chopin_short)
test = librosa.logamplitude(np.abs(chopin_stft)**2,ref_power=np.max)
librosa.display.specshow(test,y_axis='log', x_axis='time')
```
